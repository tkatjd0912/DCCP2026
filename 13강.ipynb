```python
def countdown(n: int)-> None:
    """n부터 1까지 카운트다운"""
    if n <= 0:
        print("발사!")
    else:
        print(n)
        countdown(n-1)  # 자기 자신을 호출
    
countdown(3)
```

```python
'''
def factorial_iter(n:int) -> int:
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

반복문으로 작성한 팩토리얼 함수 → result 값을 갱신하며 계산
'''

"""재귀함수로 작성한 팩토리얼 함수 → 수학적 정의에 더 부합"""
def factorial(n:int) -> int:
    if n==0:
        return 1  # 기저 조건
    return n * factorial(n-1)
    
print(factorial(4))
```

```python
"""콜스택 추적 실습"""
def factorial(n:int) -> int:
    print(f"호출: factorial({n})")
    if n == 0:
        print(f"기저 조건 도달! return 1")
        return 1
    result = n * factorial(n-1)
    print(f"반환: factorial({n}) = {result}")
    return result

print(f"결과: {factorial(3)}")
```

```python
"""피보나치 수열 재귀 구현"""
def fib(n:int) -> int:
    if n==0:
        return 0
    if n==1:
        return 1
    return fib(n-1) + fib(n-2) # 두 번의 재귀 호출 → 기저 조건 두 개 필요

for i in range(10):
    print(fib(i), end=" ")

# 피보나치 수열: fib(5)의 경우 함수가 15번 호출 → 같은 값을 여러번 중복 계산

"""피보나치 수열 - memoization 활용"""
def fib_memo(n:int, memo:dict = {}) -> int:
    if n in memo:        # 이미 계산한 적 있으면
        return memo[n]   # 저장된 값을 반환 (다시 계산 X)
    if n == 0:
        result = 0
    elif n == 1:
        result = 1
    else: 
        result = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    memo[n] = result     # 계산 결과를 저장 
    return result

print("\n======")
print(fib_memo(10))
print(fib_memo(50))
print(fib_memo(100))
```

```python
"""피보나치 수열 - 자동 메모이제이션"""

from functools import lru_cache

@lru_cache(maxsize=None)  # 함수에 캐시 기능을 덧씌운다는 의미 (클로저와 유사)
def fib_cached(n:int) -> int:
    if n==0: return 0
    if n==1: return 1
    return fib_cached(n-1) + fib_cached(n-2)

print(fib_cached(100))
```

```python
"""문자열 뒤집기 - 재귀"""
def reverse(s:str) -> str:
    if len(s) <= 1:
        return s
    return reverse(s[1:]) + s[0]

print(reverse("recursion"))
```

```python
'''
거듭제곱 - O(n)
def power_simple(x:float, n:int) -> float:
    if n==0:
        return 1
    return x * power_simple(x, n-1)

print(power_simple(2, 10))
'''



"""거듭제곱 - O(log n)"""
def power_fast(x:float, n:int) -> float:
    if n==0:
        return 1
    if n%2==0:
        half = power_fast(x, n//2)
        return half * half
    else:
        half = power_fast(x, (n-1)//2)
        return x * half * half

print(power_fast(2, 10))  # 함수를 4번만 호출 → 더 빠르다
```

```python
"""이진 탐색 - 재귀"""  # "정렬된" 데이터를 전제 
def binary_search(lst:list, target:int, 
                  low:int, high:int) -> int:
    if low > high:
        return -1
    mid = (low + high) // 2
    if lst[mid] == target:
        return mid
    elif lst[mid] < target:
        return binary_search(lst, target, mid+1, high)
    else:
        return binary_search(lst, target, low, mid-1)

numbers = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(binary_search(numbers, 23, 0, len(numbers)-1))
print(binary_search(numbers, 10, 0, len(numbers)-1))  # target이 없는 경우
```

```python
"""병합 함수"""
def merge(left:list, right:list) -> list:
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

"""병합 정렬 (재귀)"""
def merge_sort(lst:list)-> list:
    if len(lst) <= 1:  # 기저 조건 
        return lst
    mid = len(lst) // 2
    left = merge_sort(lst[:mid])
    right = merge_sort(lst[mid:])
    return merge(left, right)

data = [38, 27, 43, 3, 9, 82, 10]
print(merge_sort(data))
```

```python
"""하노이 탑"""
def hanoi(n:int, src:str, dst:str, tmp:str) -> None:
    if n==1:
        print(f"원판 1: {src} --> {dst}")
        return 
    hanoi(n-1, src, tmp, dst)
    print(f"원판 {n}: {src} --> {dst}")
    hanoi(n-1, tmp, dst, src)

hanoi(3, "A", "C", "B")
```

```python
"""실습 1: 리스트의 합"""
def recursive_sum(lst:list) -> int:
    if len(lst) == 0:
        return 0
    return lst[0] + recursive_sum(lst[1:])

print(recursive_sum([1, 2, 3, 4, 5]))
print(recursive_sum([]))
```

```python
"""실습 2: 리스트 원소 개수 세기"""
def count(lst:list) -> int:
    if lst==[]:
        return 0
    return 1 + count(lst[1:])

print(count([3, 56, 1, 34]))
```

```python
"""실습 3: 회문 검사"""
def is_palindrome(s:str) -> bool:
    if len(s) <= 1:                  # 기저 조건
        return True
    if s[0] != s[-1]:
        return False
    return (is_palindrome(s[1:-1]))  # 양끝을 제거한 뒤 재귀

print(is_palindrome("기러기"))
print(is_palindrome("언어학과"))
```

```python
"""실습 4: 중첩 리스트 평탄화"""
def flatten(lst:list) -> list:
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

data = [1, [2, 3], [4, [5, 6]], 7, [[8], 9]]
print(flatten(data))
```
