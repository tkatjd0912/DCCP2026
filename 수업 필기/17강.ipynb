```python
# 예제 1
import re
def normalize_whitespace(text: str) -> str:
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text)
    return text.strip()

normalize_whitespace("안녕  세상\n\n\n파이썬\t\t재미")
```

```python
# 예제 2
html = "<p>안녕 <b>세상</b>!</p><p>두 번째 <i>문단</i></p>"

re.sub(r"<[^>]+>", "", html)
```
