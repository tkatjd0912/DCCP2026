```python
# try and except
try:
    age = int(input("나이를 입력하세요: "))
    print(f"내년에는 {age+1}살이군요")
except ValueError:
    print("숫자를 입력해야 합니다.")
```

```python
# try and except + as
try:
    nums = [10, 20, 30]
    print(nums[100])
except IndexError as e:
    print(f"오류 발생: {e}")
    print(f"오류 종류: {type(e).__name__}")
```

```python
# 종류별로 예외 처리
def safe_divide(text_a:str, text_b:str) -> float:
    try:
        a = int(text_a)
        b = int(text_b)
        return a/b
    except ValueError:
        print("숫자만 입력하세요")
    except ZeroDivisionError:
        print("0으로는 나눌 수 없어요")

print(safe_divide("10", "2"))
safe_divide("24", "0")
safe_divide("스물", "하나")
```

```python
# 종류별로 예외 처리 - 튜플
try:
    a = int(input("a: "))
    b = int(input("b: "))
    print(a/b)
except (ValueError, ZeroDivisionError) as e:
    print(f"입력이 잘못되었습니다: {e}")
```

```python
# 예외의 가족 관계
try:
    do_something()
except Exception as e:  # 거의 모든 예외를 잡아낼 수 있다
    print(f"무슨 일이 생겼어요: {e}")
```

```python
# 예외 처리 - else: try 안에는 꼭 필요한 코드만 작성
try:
    f = open("data.txt", "r", encoding = "utf-8")
except FileNotFoundError:
    print("파일이 없어요")
else:   # except에 해당하지 않을 경우에만 코드 실행
    text = f.read()
    print(f"{len(text)}자 읽음")
    f.close()
```

```python
# 예외 처리 - finally: 반드시 실행해야 하는 코드 
f = open("data.text", "r", encoding = "utf-8")
try:
    text = f.read()
    process(text)
except OSError as e:
    print(f"읽기 실패: {e}")
finally:  # 예외 여부에 상관없이 꼭 실행해야 하는 코드 
    f.close()
```

```python
# raise: 예외를 일부러 발생시키기
def square_root(x:float) -> float:
    if x<0:
        raise ValueError(f"음수의 제곱근은 정의되지 않습니다: {x}")
    return x ** 0.5

print(square_root(-5))  # ValueError
```

```python
# class: 사용자 정의 예외
class CorpusError(Exception):
    """말뭉치 처리 중 발생하는 오류의 공통 부모."""
class EmptyFileError(CorpusError):
    """파일은 있지만 내용이 비어 있을 때."""
class InvalidEncodingError(CorpusError):
    """인코딩이 잘못된 파일."""
```

```python
# class 활용
def load_corpus(path: str) -> str:
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    if not text.strip():
        raise EmptyFileError(f"내용이 없습니다: {path}")
    return text

try:
    text = load_corpus("empty.text")
except EmptyFileError as e:      # 작은 범위의 오류 명령어를
    print(f"건너뜁니다: {e}") 
except CorpusError as e:         # 큰 범위(공통 부모)의 오류 명령어보다
    print(f"말뭉치 오류: {e}")    # 먼저 작성해야 함
except OSError as e:
    print(f"파일/시스템 오류: {e}")
```

```python
# assert: 가정을 명시하기
"""assert 조건, "조건이 거짓일 때 보일 메시지"""
def average(nums: list[float]) -> float:
    assert len(nums) > 0, "빈 리스트의 평균은 정의할 수 없음"
    return sum(nums)/len(nums)

print(average([]))  # → AssertionError 발생: 앞서 입력한 오류 메시지 출력
```

```python
# print() 디버깅
def average(nums: list[float]) -> float:
    print(f"[debug] nums = {nums}")
    total = sums(nums)
    print(f"[debug] total = {total}")
    return total / len(nums)
# → 끝나면 print()를 지워야 하며, 큰 프로그램에서는 출력이 너무 많다는 단점
```

```python
# breakpoint() 디버깅
def average(nums: list[float]) -> float:
    total = sum(nums)
    breakpoint()
    return total / len(nums)

print(average([10, 20, 30]))
```
