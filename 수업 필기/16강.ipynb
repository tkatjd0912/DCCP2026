```python
"""정규표현식"""
import re

re.findall(r"고양이", "검은 고양이와 흰 고양이")
```

```python
"""."""
re.findall(r"고.이", "고양이 고기 고래이 고래")
```

```python
"""[ ]"""
re.findall(r"[김이박]씨", "김씨와 이씨, 그리고 박씨, 최씨")
```

```python
"""[a-z]"""
re.findall(r"[가-힣]+", "Hello 안녕하세요 WOrld 반갑습니다 12345")
```

```python
"""부정 ^"""
re.findall(r"[^가-힣]+", "안녕 World 반갑 2026")
```

```python
"""단축 클래스"""
text = "주문번호: 2026-0413, 가격: 12,000원, 연락처: 010-1234-5678"
print(re.findall(r"\d+", text))
print(re.findall(r"\S+", "안녕\t세상\t파이썬"))
```

```python
"""^, $ 시작과 끝"""
print(re.findall(r"^안녕", "안녕 세상"))
print(re.findall(r"^세상", "안녕 세상"))
```

```python
"""\b 단어 경계"""
re.findall(r"\bcat\b", "cat, scatter, cathedral, the cat sat")
```

```python
"""수량자: ?(있으나 마나), *(0개 이상), +(1개 이상)"""
print(re.findall(r"colou?r", "color and colour"))
print(re.findall(r"go+gle", "ggle gogle google goooooogle"))
print(re.findall(r"\d{3}-\d{4}", "전화: 880-6163"))
```

```python
"""탐욕적 매칭"""
re.findall(r"<.+>", "<b>굵게</b> 그리고 <i>기울임</i>")
```

```python
"""게으른 매칭"""
re.findall(r"<.+?>", "<b>굵게</b> 그리고 <i>기울임</i>")
```

```python
"""그룹화(캡쳐)"""
m = re.search(r"(\d{4})-(\d{2})-(\d{2})", "오늘은 2026-05-04")
print(m.group(0))
print(m.group(1))
print(m.group(2))
print(m.group(3))
print(m.groups())
```

```python
"""| (or)"""
print(re.findall(r"강아지|고양이", "우리 집 고양이와 옆집 강아지"))
print(re.findall(r"(아침|점심|저녁) 식사", "아침 식사 후 점심 식사"))
```

```python
"""그룹화(비캡쳐)"""
m = re.search(r"(?:Mr|Ms|Dr)\.(\w+)", "Hello Dr.Park")
m.group(1)
```

```python
"""그룹화(명명)"""
m = re.search(
    r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", 
    "오늘은 2026-05-04"
)

print(m.group("year"))
print(m.groupdict())
```

```python
"""역참조"""
re.sub(r"(\d{4})-(\d{2})-(\d{2})", r"\3/\2/\1", "2026-05-04")
```

```python
"""Lookahead"""
text = "가격은 1500원, 무게는 2300g, 거리는 500m"
print(re.findall(r"\d+(?=원)", text))  # Positive Lookahead
print(re.findall(r"\d+(?!원)", text)) # Negative Lookahead
```

```python
"""Lookbehind"""
text1 = "가격은 $100, 할인 10%, $250 행사"
text2 = "한국, 미국, 영국, 중국, 미군, 미주"
print(re.findall(r"(?<=\$)\d+", text1))  # Positive Lookbehind
print(re.findall(r"(?<!미)국", text2))   # Negative Lookbehind
```
