```python
# 텍스트 파일 생성
with open ('poem.txt', 'w', encoding='utf-8') as f:
    f.write("죽는 날까지 하늘을 우러러\n")
    f.write("한 점 부끄럼이 없기를\n")
    f.write("잎새에 이는 바람에도\n")
    f.write("나는 괴로워했다\n")
```

```python
# 읽기 
f = open("poem.txt", "r", encoding='utf-8')
text = f.read()
f.close()
print(text)
print(f"글자 수: {len(text)}")
```

```python
# 한 줄씩 처리 
f = open("poem.txt", "r", encoding='utf-8')
for i, line in enumerate(f, start=1):
    print(f"{i}행: {line.rstrip()}")
f.close()
```

```python
# with 활용
with open("poem.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        line = line.rstrip()
        print(f"{i}행 ({len(line)}자): {line}")
```

```python
# 단어 빈도 계산
text = "the cat sat on the mat the cat slept"
words = text.split()
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

with open("freq.txt", "w", encoding="utf-8") as f:
    for word, count in sorted(freq.items()):
        f.write(f"{word}\t{count}\n")
```

```python
# 헤더 분리
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)
    print(f"열 이름: {header}")
    for row in reader:
        name, sid, dept = row   # 언패킹
        print(f"{name} ({sid}) - {dept}")
```

```python
# 행을 딕셔너리로 
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row["이름"], row["학과"])
```

```python
# 행을 리스트로 쓰기
import csv

rows = [
    ["언어", "사용자수(백만)", "어족"],
    ["한국어", 81, "한국어족"],
    ["영어", 1500, "인도유럽어족"],
    ["스페인어", 600, "인도유럽어족"]
]

with open("languages.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)      # newline: 빈 줄이 끼어드는 것을 방지
    writer.writerows(rows)
```
